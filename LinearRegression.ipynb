**Solution for question 1**

```python
x1 = np.random.normal(5,7,1000)
```

```python
sns.distplot(x1,kde=False, fit=st.norm)
```

```python
df = pd.DataFrame(dict(x=np.arange(len(x1)),x1=x1))
```

```python
sns.relplot(x="x",y="x1",data=df)
```

```python
true_error = np.random.normal(0,2,1000)
```

```python
true_beta_0 = 1.1
true_beta_1 = -8.2
```

```python
y = true_beta_0 + true_beta_1*x1 + true_error
```

```python
df_y = pd.DataFrame(dict(x=np.arange(len(y)),y=y))
```

```python
sns.relplot(data=df_y)
```

```python
sns.distplot(y, kde=False, fit=st.norm)
```

```python
x1 = np.reshape(x1,(1000,1))
y = np.reshape(y, (1000,1))
```

```python
c = np.ones((1000,1))
```

```python
A = np.hstack([x1,c])
```

```python
np.linalg.lstsq(A,y,rcond=None)
```
